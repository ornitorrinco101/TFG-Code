```r
library(latex2exp)
```

```r
data1=read.table("/Users/mcytron/Desktop/TFG/CODE/lepton-nucleon-c++/Ei1000.out",header=FALSE)
data2=read.table("/Users/mcytron/Desktop/TFG/CODE/lepton-nucleon-c++/Ei500.out",header=FALSE)
names(data1)=c("omega","K_f","P_n","Q2","dsigQ2","costheta_f","dsig")
names(data2)=c("omega","K_f","P_n","Q2","dsigQ2","costheta_f","dsig")
#data$Q es omega
```

```r
plot(acos(data1$costheta_f)*180/pi,data1$dsig,
     xlab="\\theta\\;(deg)",
     ylab=" ",
     cex.lab=0.75,cex.axis=0.75,
     type="l",
     col="darkorchid")

lines(acos(data2$costheta_f)*180/pi,data2$dsig,
      xlab="$\\theta\\;(deg)$",
      ylab="",
      cex.lab=0.75,cex.axis=0.75,
      type="l",
      col="deeppink")

legend(x="topright",
       legend=c("\\epsilon_{i}=1000\\;MeV$","$\\epsilon_{i}=500\\;MeV"),
       col=c("darkorchid","deeppink"),lty=c("solid","solid"),cex=0.6)

mtext("$d\\sigma/d\\cos(\\theta)\\;\\;(fm^{2})$",side=2,line=2,cex=0.75)
mtext("$Sección\\;eficaz\\;diferencial\\;del\\;proceso\\;\\nu_{\\mu}+n\\to\\mu^{-}+p",side=3,las=1,line=0.8,cex=1)
```

```r
plot(data1$Q2,data1$dsigQ2,
     xlab="",
     ylab="",
     cex.lab=0.75,cex.axis=0.75,
     type="l",
     col="darkorchid")

mtext("$Q^{2}\\;(GeV^{2})$",side=1,line=2,cex=0.75)
mtext("$d\\sigma(Q^{2})/dQ^{2}\\;(fm^{2}/MeV^{2})$",side=2,line=2,cex=0.75)
mtext("$Sección\\;eficaz\\;diferencial\\;del\\;proceso\\;\\nu_{\\mu}+n\\to\\mu^{-}+p$",side=3,las=1,line=1,cex=1)

legend(x="topright",
       legend=c("$\\epsilon_{i}=1\\;GeV$"),
       col=c("darkorchid"),lty=c("solid"),cex=0.6)

```

```r
plot(data2$Q2,data2$dsigQ2,
     xlab="",
     ylab="",
     cex.lab=0.75,cex.axis=0.75,
     type="l",
     col="deeppink")

mtext("$Q^{2}\\;(GeV^{2})$",side=1,line=2,cex=0.75)
mtext("$d\\sigma(Q^{2})/dQ^{2}\\;(fm^{2}/MeV^{2})$",side=2,line=2,cex=0.75)
mtext("$Sección\\;eficaz\\;diferencial\\;del\\;proceso\\;\\nu_{\\mu}+n\\to\\mu^{-}+p$",side=3,las=1,line=1,cex=1)

legend(x="topright",
       legend=c("$\\epsilon_{i}=500\\;MeV$"),
       col=c("deeppink"),lty=c("solid"),cex=0.6)
```
